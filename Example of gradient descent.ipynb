```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# целевая функция
def func(x):
    return np.square(x)
```

```python
# производная целевой функции
def derivative(x):
    return 2 * x
```

```python
x0 = 10 # Начальная точка
learning_rate = 0.1 # скорость шага градиентного спуска
epoch = 100
```

```python
x_array = []
y_array = []

for _ in range(epoch):
    dx = derivative(x0) # находим производную
    x0 = x0 - learning_rate * dx # обновляем x0
    
    x_array.append(x0)
    y_array.append(func(x0))
```

```python
x = np.linspace(-9, 9, 100)
y = func(x)
```

```python
plt.figure(figsize=(12, 8), dpi=200)
plt.plot(x, y, color='green')
plt.fill(x, y, alpha=0.2, color='purple')
plt.scatter(x_array, y_array, color='red')

plt.xlabel('x')
plt.ylabel('y');
```
